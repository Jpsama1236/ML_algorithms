```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=5, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#using gradient descent approach and keeping m as fixed , we will vary b,lets say b=100
```

```python
# from above model we got m=74.25, b=7.79 , we will try to reach them by gradient descent approach
#iteration 1
m=74.25
b=100
ypred=(m*X)+b
ypred=ypred.reshape(5)
```

```python
plt.scatter(X,y)
plt.plot(X,ypred,color='green',label='b = {}'.format(b))
plt.plot(X,lr.predict(X),color='red')
plt.legend()
```

```python
#now we need to find slope at b
slope=-2*np.sum(y-m*X.ravel()-b)
slope
step_size=slope*0.08
step_size
b1=b-step_size
b1
```

```python
#iteration 2
m=74.25
ypred1=((m*X)+b1).reshape(5)
```

```python
plt.scatter(X,y)
plt.plot(X,ypred1,color='blue',label='b = {}'.format(b1))
plt.plot(X,ypred,color='green',label='b = {}'.format(b))
plt.plot(X,lr.predict(X),color='red')
plt.legend()
plt.show()
```

```python
#now we need to find slope at b
slope=-2*np.sum(y-m*X.ravel()-b1)
slope
step_size=slope*0.08
step_size
b2=b1-step_size
b2
```

```python
#iteration 3
m=74.25
ypred2=(m*X)+b2
```

```python
plt.scatter(X,y)
plt.plot(X,ypred,color='green',label='b = {}'.format(b))
plt.plot(X,ypred1,color='blue',label='b = {}'.format(b1))
plt.plot(X,ypred2,color='#A3E4D7',label='b = {}'.format(b2))
plt.plot(X,lr.predict(X),color='red')
plt.legend()
plt.show()
```

```python
#we can write it in loop
b = -100
m = 74.25
lrate = 0.06

epochs = 10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lrate * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred,label='b = {}'.format(b))

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
plt.legend()
plt.show()
```
