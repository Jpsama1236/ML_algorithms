```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_features=5,n_informative=5,n_redundant=0,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
df.head()
```

```python
def sample_rows(df,percent):
  return df.sample(int(percent*df.shape[0]),replace=True)
```

```python
import random
```

```python
def sample_features(df,percent):
  cols=df.columns.to_list()[:-1]
  n=int(percent*df.shape[1]-1)
  df_new=df[random.sample(cols,int(percent*df.shape[1]-1))]
  df_new['target']=df['target']
  return df_new
```

```python
df.columns.to_list()
```

```python

# function for combined sampling

def combined_sampling(df,row_percent,col_percent):
  new_df = sample_rows(df,row_percent)
  return sample_features(new_df,col_percent)
```

```python
df1=sample_rows(df,0.5)
```

```python
df1=sample_features(df,0.5)
```

```python
df1.head()
```

```python
df1=combined_sampling(df,0.5,0.5)
```

```python
df1.shape
```

```python
df2=sample_features(df,0.5)
```

```python
df3=sample_features(df,0.5)
```

```python
df3
```

```python

from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
clf1.fit(df1.iloc[:,0:df1.shape[1]-1],df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:df2.shape[1]-1],df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:df3.shape[1]-1],df3.iloc[:,-1])
```

```python

from sklearn.tree import plot_tree
plot_tree(clf1)
```
